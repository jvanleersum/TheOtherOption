```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../TheOtherOption/data/all_products_ah_nutr_allergies.csv")
```

```python
df.info()
```

```python
df["features"] = df["features"].fillna('no_info')
```

```python
all_features = []
for features in df["features"]:
    if features != "no_info":
        for f in eval(features):
            all_features.append(f)
all_features = set(all_features)
```

```python
all_features
```

```python
features_df = pd.DataFrame(all_features)
features_df.to_csv("../TheOtherOption/data/features_list.csv", index=False)
```

```python
for feature in all_features:
    df[feature] = ""
```

```python
for ind, feature in df["features"].items():
    if feature != "no_info":
        for f in eval(feature):
            df.loc[ind, f] = "feature"
```
