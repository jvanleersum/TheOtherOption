```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("../TheOtherOption/data/all_products_ah_nutr_allergies.csv")
```

```python
df.head()
```

```python
df["features"] = df["features"].fillna('no_info')
```

```python
all_features = []
for features in df["features"]:
    if features != "no_info":
        for f in eval(features):
            all_features.append(f)
all_features = set(all_features)
```
