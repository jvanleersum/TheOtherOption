```python
features = df_features["0"].tolist()
features.insert(0, "selecteer een feature")
features
```

```python
categories = df_categories["0"].tolist()
categories.insert(0, "selecteer een categorie")
categories
```

```python
allergies = df_allergies["0"].tolist()
allergies.insert(0, "selecteer een allergie")
allergies
```

```python
from ipywidgets import interact, Dropdown
select_features = Dropdown(options = features)
select_categories = Dropdown(options = categories)
select_allergies = Dropdown(options = allergies)
```

```python
@interact(feature = select_features, category = select_categories)
def display_feature(feature):
    if feature == "selecteer een feature":
        return(df_products)
    else:
        return(df_products[df_products[feature] == "feature"])
```

```python
@interact(category = select_categories)
def display_category(category):
    if category == "selecteer een categorie":
        return(df_products)
    else:
        return(df_products[df_products[category] == "category"])
```

```python
@interact(allergy = select_allergies)
def display_allergy(allergy):
    if allergy == "selecteer een allergie":
        return(df_products)
    else:
        return(df_products[df_products[allergy].apply(lambda x:x not in ["allergy", "kan bevatten"])])
```

```python
product = input("Voer een product in")
def display_product(product):
    if product == "":
        return df_products
    elif product in categories:
        return df_products[df_products[product].apply(lambda x:x in ["category"])]
    else:
        return print(f"Categorie {product} niet gevonden")
display_product(product)
```
