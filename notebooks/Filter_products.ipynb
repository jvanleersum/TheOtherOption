```python
from ipywidgets import interact, Dropdown
select_features = Dropdown(options = features)
select_allergies = Dropdown(options = allergies)
if select_main_categories == "selecteer een hoofdcategorie":
    select_sub_categories = Dropdown(options = sub_categories)
else:
    select_sub_categories = Dropdown(options = set(df_products[df_products["main_category"]\
                                                               == select_main_categories]["sub_category"]))
```

```python
if select_main_categories == "selecteer een hoofdcategorie":
    select_sub_categories = sub_categories
else:
    select_sub_categories = set(df_products[df_products["main_category"] == select_main_categories]["sub_category"])
    
print(select_sub_categories)
```

```python
@interact(feature = select_features)
def display_feature(feature):
    if feature == "selecteer een feature":
        return(df_products)
    else:
        return(df_products[df_products[feature] == "feature"])
```

```python
@interact(main_category = select_main_categories, sub_category = select_sub_categories)
def display_category(main_category, sub_category):
    if (main_category == "selecteer een hoofdcategorie") & (sub_category == "selecteer een subcategorie"):
        return(df_products)
    elif (main_category == "selecteer een hoofdcategorie"):
        return(df_products[df_products["sub_category"] == sub_category])
    elif (sub_category == "selecteer een subcategorie"):
        return(df_products[df_products["main_category"] == main_category])
    else:
        return(df_products[(df_products["main_category"] == main_category) & 
                                  (df_products["sub_category"] == sub_category)])
```

```python
@interact(allergy = select_allergies)
def display_allergy(allergy):
    if allergy == "selecteer een allergie":
        return(df_products)
    else:
        return(df_products[df_products[allergy].apply(lambda x:x not in ["allergy", "kan bevatten"])])
```
